```python
import matplotlib.pyplot as plt
import numpy as np
from simulation import monte_carlo_simulation
```

```python
num_simulations = 1000
step = 50
rows, seats_per_row, exits = 30, 6, [0, 15, 29]

avg_times = []
for i in range(step, num_simulations + 1, step):
    evacuation_times = monte_carlo_simulation(rows=rows, seats_per_row=seats_per_row, exits=exits,
                                              num_simulations=i)
    avg_times.append(np.mean(evacuation_times))

# Plot results
plt.plot(range(step, num_simulations + 1, step), avg_times, marker='o')
plt.xlabel('Number of Simulations')
plt.ylabel('Average Evacuation Time')
plt.title('Convergence of Monte Carlo Simulation')
plt.grid()
plt.show()
```
